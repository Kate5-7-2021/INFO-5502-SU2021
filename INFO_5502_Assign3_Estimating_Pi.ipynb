##**Estimating π using the Monte Carlo Method**

Write a Python program to simulate the calculation of π using Monte Carlo method. Your program draws the circle and the square, and plot the random points in the graph as shown in Figure 1. You need simulate different numbers of points (at least include 10, 500, 1,000, 5,000, 10,000, 50,000, 80,000, 100,000, 200,000) and calculate the value π for each different number of points. Then draw a line chart to show the change of the accuracy of the value of π calculated when the number of points is increased.


```python
import random                 #import necessary libraries
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
random.seed(10000)            #define number of the random seed
def estimate(N_total):
  N_inner =                   #define the inner points

  for i in range(0,N_total):   #loop through different total values to calculate the Pi value 
    x=random.random()**2        #define random x & y
    y=random.random()**2

    if sqrt(x+y) <=1:          
      N_inner += 1

  pi = (4*N_inner)/N_total      #calculate Pi values based on the total & inner points
  return pi 

N_total=[10,500,1000,5000,10000,50000,80000,100000,200000]   #define number of the total points


pi_dict= {}                                       #create dictionary
for j in N_total:
  pi_dict[j]=estimate(j)                          #estimate the Pi related to each total number

for k in (N_total):
  print('total points:' +str(k) +'  ', 'estimate of pi:' + str(pi_dict[k]))  # print the pi value related to each total number
#for l in N_total:
  #x1=numpy.random.uniform(-1,1)
  #y1=numpy.random.uniform(-1,1)
  #if numpy.sqrt(x**2 +y**2)<1:
    

```
